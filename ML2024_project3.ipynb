```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('winequality-white.csv', delimiter=';')
df_copy = df.copy()

# Standardize the data
scaler = StandardScaler()
features = df_copy.drop('quality', axis=1)  # Exclude the target variable 'quality'
x = scaler.fit_transform(features)

# Calculate the within-cluster sum of squares (inertia) for different numbers of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
# prompt: Experiment with different values for `n_clusters` and see if you can find a better score.

# Try different numbers of clusters
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)

    # Calculate the confusion matrix
    labels = kmeans.labels_
    df_clusters = pd.DataFrame({'cluster': labels, 'quality': df['quality']})
    confusion_matrix = pd.crosstab(df_clusters['cluster'], df_clusters['quality'])

    # Print the score
    print(f"k={k}:")
    print(confusion_matrix)

```

```python
# prompt: calculate accuracy score for these k tested above

from sklearn.metrics import accuracy_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)

    # Calculate the confusion matrix
    labels = kmeans.labels_
    df_clusters = pd.DataFrame({'cluster': labels, 'quality': df['quality']})
    confusion_matrix = pd.crosstab(df_clusters['cluster'], df_clusters['quality'])

    # Calculate the accuracy score
    accuracy = accuracy_score(df_clusters['quality'], labels)

    # Print the score
    print(f"k={k}:")
    print(f"Accuracy: {accuracy}")

```

```python
# Determine the optimal number of components using the elbow method
# Calculate the AIC and BIC

from sklearn.mixture import GaussianMixture
aics = []
bics = []
n_components_range = range(1, 11)
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(x)
    aics.append(gmm.aic(x))
    bics.append(gmm.bic(x))

# Plot the elbow plots for AIC and BIC
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(n_components_range, aics, marker='o')
plt.title('GMM - AIC')
plt.xlabel('Number of components')
plt.ylabel('AIC')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(n_components_range, bics, marker='o')
plt.title('GMM - BIC')
plt.xlabel('Number of components')
plt.ylabel('BIC')
plt.grid(True)

plt.show()
```

```python
# prompt: what is the best n based on the plots above

Based on the plots above, the best n is **3** according to the elbow method for both AIC and BIC.
```

```python
# prompt: How much variation did you get when you re-ran your random projections
# several times?

The amount of variation obtained when re-running random projections several times can be quantified using various metrics, such as the mean squared error (MSE) or the root mean squared error (RMSE) between the original data and the reconstructed data after each projection.

Here's how to calculate the MSE for multiple runs of random projections:

```

```python
from sklearn import random_projection
from sklearn.metrics import mean_squared_error

n_components = min(x.shape[1] - 1, 10)
transformer = random_projection.GaussianRandomProjection(n_components=n_components)

mse_values = []
n_runs = 10  # Number of runs

for _ in range(n_runs):
    # Create a new random projection
    X_projected = transformer.fit_transform(data_attributes_std)
    # Inverse transform/reconstruct the original data
    X_reconstructed = transformer.inverse_transform(X_projected)
    # Calculate MSE
    mse = mean_squared_error(data_attributes_std, X_reconstructed)
    mse_values.append(mse)

# Calculate the average MSE over all runs
average_mse = np.mean(mse_values)
print(f'Average MSE over {n_runs} runs: {average_mse}')
std_dev_mse = np.std(mse_values)
variance_mse = np.var(mse_values)
print(f'variance MSE over {n_runs} runs: {variance_mse}')
```

```python
#Linear Discriminant Analysis (LDA) --another linear dimention reduction algorithm

# Create and fit LDA model
lda = LinearDiscriminantAnalysis()

lda.fit(X_train, y_train)

# Reduce dimensions of train and test data
X_test_reduced = lda.transform(X_test)
X_train_reduced = lda.transform(X_train)

# Fit the MLPClassifier on the reduced data

start_time = time()
clf.fit(X_train_reduced, y_train)
end_time = time()


# Make predictions on train and test data
prediction_train = clf.predict(X_train_reduced)
prediction_test = clf.predict(X_test_reduced)
train_accuracy = accuracy_score(y_train, prediction_train)
test_accuracy = accuracy_score(y_test, prediction_test)

print(f"Training Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
print(f"Training Time: {end_time - start_time} seconds")
```
